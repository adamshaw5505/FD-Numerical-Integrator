```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

def plot_constraints(name,mods):
    if len(mods) == 0: return
    files = []
    legends = []
    file_string = "./Data/"+name
    file_name_list = []
    lengeds = []
    for mod in mods:
        file_string_temp = file_string[:]
        for key in mod.keys():
            file_string_temp += "-"+  key +"-"+mod[key]
            if key == "N":
                legends.append(mod[key])
        file_string_temp += "-Const.dat"
        file_name_list.append(file_string_temp)
        
    files = [open(file_str) for file_str in file_name_list]
    
    datas = [np.array([[float(num) for num in line.split(", ") if num != ", "] for line in file.readlines()]) for file in files]

    for i in range(datas[0].shape[1]-1):
        fig,ax = plt.subplots()
        for data in datas:
            ax.plot(data[:,0],np.log(data[:,i+1]))
        ax.set_xlabel("Simulation Time")
        ax.set_ylabel(r'$\log_{10}(\Delta u(t))$')
        ax.set_title(name + " $G_"+str(i)+"$")
        ax.legend(legends)
```
